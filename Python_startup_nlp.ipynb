## 自然言語処理、独特の処理をしてみよう
当然ながら自然言語は数字ではありません。
なので数字を扱う機械学習モデルに、なんの工夫なしに突っ込むことはできません。<br>
そこで、自然言語を機械学習モデルで扱えるように加工のする必要があります。<br>
今回は**BoW (Bag of Words)**という方法で加工して、<br>
機械学習モデルに突っ込んでみましょう。<br>
<br>
BoW(Bag of Words)とは……

1. 文章をトークンを作成
1. 語彙を作成する
1. 各文章で各単語の出現回数をカウントする。

はじめは簡単な文章から加工してみましょう。

```
text = np.array(["I will teach you justice with my stand",
                 "You say good bye , I say hello",
                 "Bring it on , if you want to die"])
```

```
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
bag = count.fit_transform(text)

print(count.vocabulary_)
print("-"*50)
print(bag.toarray())
```

ここにきて3分間クッキング方式を取ります。<br>
今回はKeras.datasetsよりIMDbデータセットを拝借。<br>
手軽に実行できるのでこちらを採用。<br>
はっはっは、安定してますな。<Br>

```
from keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)
```

```
print(x_test[0])
```

```
print("x_train shape : ", x_train.shape)
print("y_train shape : ", y_train.shape)
print("-"*60)
print("x_test shape : ", x_test.shape)
print("y_test shape : ", y_test.shape)
```

```
y_test
```

```
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

x_train = vectorize_sequences(x_train)
x_test = vectorize_sequences(x_test)

x_train[0]
```

```
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```
print(clf.predict(x_test))
```

```
print(clf.score(x_test, y_test))
```

## Step2 学習データの準備
機械学習ではデータが無くては何も出来ません。<br>
今回は学習データとして使用するニーチェの文章を、<br>
下記のセルで準備しています。

```
import keras
import numpy as np

path = keras.utils.get_file(
    'nietzsche.txt',
    origin='https://s3.amazonaws.com/text-datasets/nietzsche.txt')
text = open(path).read().lower()
print('Corpus length:', len(text))
```

## Step3  データの前処理
ここではDeep Learningで使用出来るように加工しています。<br>
機械学習ではこの様な行程を前処理と呼びます。<br>
この前処理といわれる行程は、<br>
機械学習の作業の大半を占めている非常に重要な作業です。<br>
今回の行程を具体的に覚えておく必要はありませんが、<br>
**前処理はクソ面倒かつ、重要な作業である**ことは覚えておきましょう。<br>
<br>
今回行われている具体的な処理としては……
1.
1.
1.


```
# Length of extracted character sequences
maxlen = 60

# We sample a new sequence every `step` characters
step = 3

# This holds our extracted sequences
sentences = []

# This holds the targets (the follow-up characters)
next_chars = []

for i in range(0, len(text) - maxlen, step):
    sentences.append(text[i: i + maxlen])
    next_chars.append(text[i + maxlen])
print('Number of sequences:', len(sentences))

# List of unique characters in the corpus
chars = sorted(list(set(text)))
print('Unique characters:', len(chars))
# Dictionary mapping unique characters to their index in `chars`
char_indices = dict((char, chars.index(char)) for char in chars)

# Next, one-hot encode the characters into binary arrays.
print('Vectorization...')
x = np.zeros((len(sentences), maxlen, len(chars)), dtype=np.bool)
y = np.zeros((len(sentences), len(chars)), dtype=np.bool)
for i, sentence in enumerate(sentences):
    for t, char in enumerate(sentence):
        x[i, t, char_indices[char]] = 1
    y[i, char_indices[next_chars[i]]] = 1
```

## Step4 モデル作成
おまたせしました。<br>
今回の処理で使用するDeep Learning のモデルを作成します。<br>
Kerasでの書き方はとっても簡単。<br>
<br>
自分が作成するモデルの各層をリスト型として纏めるだけ。<br>
これでチャチャっとDeepLearning出来るのだから驚き。<br>
とっても簡単だから、みんなもDeep Learningしようぜ！<br>

```
from keras import layers

model = keras.models.Sequential()
model.add(layers.LSTM(128, input_shape=(maxlen, len(chars))))
model.add(layers.Dense(len(chars), activation='softmax'))
```

## Step5 Optimizer と lossを決定する
今日は……直感で大丈夫です。<br>
ザックリ説明するとDeepLearningを学習するにあたって<br>

<br>
ここで重要なのは...<br>

```
optimizer = keras.optimizers.RMSprop(lr=0.01)
model.compile(loss='categorical_crossentropy', optimizer=optimizer)
```

## Step6 学習開始！
ここから学習の開始です！


```
def sample(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

```
import random
import sys

for epoch in range(1, 60):
    print('epoch', epoch)
    # Fit the model for 1 epoch on the available training data
    model.fit(x, y,
              batch_size=128,
              epochs=1)

    # Select a text seed at random
    start_index = random.randint(0, len(text) - maxlen - 1)
    generated_text = text[start_index: start_index + maxlen]
    print('--- Generating with seed: "' + generated_text + '"')

    for temperature in [0.2, 0.5, 1.0, 1.2]:
        print('------ temperature:', temperature)
        sys.stdout.write(generated_text)

        # We generate 400 characters
        for i in range(400):
            sampled = np.zeros((1, maxlen, len(chars)))
            for t, char in enumerate(generated_text):
                sampled[0, t, char_indices[char]] = 1.

            preds = model.predict(sampled, verbose=0)[0]
            next_index = sample(preds, temperature)
            next_char = chars[next_index]

            generated_text += next_char
            generated_text = generated_text[1:]

            sys.stdout.write(next_char)
            sys.stdout.flush()
        print()
```
